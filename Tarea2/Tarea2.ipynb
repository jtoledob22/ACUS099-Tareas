## 1. Introducción <a name='id1'></a>

La convolución es una operación matemática fundamental en el análisis y procesamiento de señales que tiene aplicaciones en una variedad de campos, desde la ingeniería hasta la física. En términos generales, la convolución permite combinar dos señales para formar una tercera, proporcionando una forma poderosa de entender cómo una señal se modifica al pasar por un sistema. Este concepto es esencial en el diseño y análisis de sistemas lineales e invariantes en el tiempo (LTI), ya que describe cómo una señal de entrada se transforma en una señal de salida a través de la respuesta del sistema.

En el ámbito de las señales acústicas, la convolución adquiere un papel crucial. Ya que, al aplicar la convolución en estas señales, es posible modelar cómo se comportan en diferentes entornos y bajo diferentes condiciones. Por ejemplo, se puede simular cómo una habitación afecta el sonido producido en su interior, creando reverberaciones y eco que alteran la percepción auditiva.

## 1. Función delta (Señal impulso) <a name='id2'></a>
$$
\delta(x) =
\left \{
  \begin{array}{lcc}
    0 & ; &  x≠0\\
    1 & ; & x=0
  \end{array}
\right.
$$

## 2. Sistemas en Tiempo Discreto <a name='id2'></a>

Se puede definir como una transformación que modifica una secuencia de valores de entrada x[n] en una secuencia de salida con valores y[n]. Se puede expresar de la siguente manera:

$$
\begin{equation}
y[n] = T\{x[n]\}
\end{equation}
$$





### 2.1 Sistemas Lineales <a name='id2.1'></a>

Un sistema es lineal cuando cumple con 2 propiedades, la de aditividad:

$$
T\{x_1[n] + x_2[n]\} = T\{x_1[n]\}+T\{x_2[n]\} = y_1[n] + y_2[n]
$$

y con la propiedad de homogeneidad: 

$$
T\{a x[n]\} = a T\{x[n]\}= a y[n] \quad ; \quad  a \in \mathbb{R}
$$

Siendo $y_1[n]$ e $y_2[n]$ las respuestas de un sistema, cuando $x_1[n]$ y $x_2[n]$, son las respectivas entradas. Estas propiedades se pueden combinar, quedando lo siguente:

$$
T\{a x_1[n] + bx_2[n] \} = aT\{x_1[n]\} + bT\{x_2[n]\} \quad ; \quad  a, b \quad constantes
$$

### 2.2 Sistemas Invariantes con el tiempo <a name='id2.2'></a>

Se define como un sistema en el cual un retardo en la entrada, provoca el mismo retardo en la salida. Es decir que:

$$
\begin{align}
    x_1[n] & = x[n-t_0] \\
    y_1[n] & = y[n-t_0]
\end{align}
$$

Donde $t_0$ seria el retardo.

### 2.3 Sistemas Lineales e Invariantes con el tiempo (LTI) <a name='id2.3'></a>

Los sistemas LTI son aquellos sistemas que cumplen con las dos propiedades: 

1) Linealidad
2) Invarianza en el tiempo.




## 3. Suma de Convolución <a name='id3'></a>

Cualquier señal en tiempo discreto se puede escribir como una suma infinita de deltas unitarios:

$$
\begin{align}
x[n] &= \sum_{k=0}^{\infty} x[k] \cdot \delta[n - k]
\end{align}
$$

## 4. Solución Análitica Ejemplo <a name='idej'></a>
**(Ejemplo 2.13 (Oppenheim & Schafer, 1989))**
> + Consideremos un sistema con respuesta impulso es:
$$
\begin{align*}
h[n] &= u[n]-u[n-N] \\
& =\begin{cases}
1 & \text{para } 0\leq n \leq N-1 \\
0 & \text{otro caso}
\end{cases} 
\end{align*}
$$
> donde $u[n]$ es un escalón unitario centrado en cero.
>
> La entrada es:
$$x[n]=a^{n}u[n]$$
>
> <img src="fotos/fig1.png" style="height: 200px; display: block;margin-left: auto;margin-right: auto;"/>
>
> Analizando la figura (a), si utilizamos un valor de $n < 0$, la entrada y la respuesta al impulso, no se solapan.
>
> $$\therefore y[n]=0, \qquad n<0$$
>
> <img src="fotos/fig2.png" style="height: 200px; display: block;margin-left: auto;margin-right: auto;"/>
>
> Luego, como se puede observar en la figura (b) (entre $0\leq n \leq N-1$):
>
> $$
 \begin{align}
     x[k]h[n-k] & =a^k
    y[n] &= \sum_{k=0}^{n} a^k, \qquad \text{para } 0\leq n \leq N-1
 \end{align}
 $$
>  Que luego se puede expresar de la siguente manera:
> $$
\begin{align*}
\sum_{k=N_{1}}^{N_{2}}\alpha^{k} &= \frac{\alpha^{N_{1}}- \alpha^{N_{2}+1}}{1-\alpha},\quad N_{2}\geq N_{1}\\
\therefore y[n] & = \frac{1-a^{n+1}}{1-a}, \qquad 0 \leq n \leq N-1
\end{align*}
$$
>
> 
> <img src="fotos/fig3.png" style="height: 200px; display: block;margin-left: auto;margin-right: auto;"/>
>
> Por ultimo la figura (c), no muestra cuando $n < N-1$, utilizamos lo mismo que antes, pero ahora cambiando el limite inferior a $n-N+1$.
>
>  $$
 \begin{align}
     x[k]h[n-k] & =a^k, \qquad n-N+1<k\leq n \\
    y[n] &= \sum_{k=n - N+1}^{n} a^k, \qquad \text{para } N-1 < n
 \end{align}
 $$
>
> Aplicando lo anterior:
> \begin{align*}
y[n] & = \frac{a^{n-N+1}-a^{n+1}}{1-a} \\
\therefore y[n] & = a^{n-N+1} \left( \frac{1-a^N}{1-a} \right), \qquad N-1 \leq n
\end{align*}
>
> Por lo tanto:
> $$
\begin{align*}
y[n] =
\begin{cases}
0, & n<0 \\
\frac{1-a^{n+1}}{1-a}, & 0 \leq n \leq N-1 \\
a^{n-N+1} \left( \frac{1-a^N}{1-a} \right), & N-1 \leq n
\end{cases} 
\end{align*}
$$


```python
#Importamos librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

# Parámetros
a = 0.8  
N = 5    # Ancho de Escalon

def escalon(n):
    return 1*(n>=0)

def delta(n, p):
    return 1*(n==p) ## Ingresa 1 en posicion p
    
# Definimos x[k]
def signal1(x):
    return a**x * escalon(x)
    
#Definimos h[n-k]
def signal2(x):
    return np.roll(escalon(x) - escalon(x-5), 1)

# Definimos solucion analitica problema
def conv_calc(n, a, N):
    #n = np.array()

    if n < 0:
        return 0
    elif 0 <= n <= N - 1:
        return (1 - a**(n+1)) / (1 - a)
    else:
        return a**(n - N + 1) * (1 - a**N) / (1 - a)

def convpos(n, p):
    return conv_calc(p, a, N) * delta(n, p)

# Parametros señales

n = np.arange(-5,16)
y1 = signal1(n)
y2 = signal2(n)
```

```python
#y1 = (1-a**(n+1))/(1-a)
#y2 = a**(n-N+1)*((1-a**N)/(1-a))

x = a**n * escalon(n)
h = escalon(n) - escalon(n-N)
#h2 = escalon(n-N)

conv_analitica=[]

for i in range(-5,16):
    conv_analitica.append(conv_calc(i,a,N))

fig1, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6), layout="constrained")

# Convolucion Analitica
ax1.stem(n,conv_analitica, linefmt='red');
ax1.set_xlim(-5,15)
ax1.set_title('Solución Analítica')


ax2.stem(np.roll(np.convolve(x,h),-10), linefmt='black');
ax2.set_xlim(-5,15)
ax2.set_title('Solución Numpy');
```

```python
# Gráfico de las señales
fig2, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6), layout="constrained")

ax2.set_xlim(-5, 15)
ax2.set_ylim(0, 4)
ax2.set_title('Convolución')

# Función de actualización para la animación
def update(frame):

    t = frame - 10

    y2_shifted = signal2(n-t)

    ax1.cla()

    #Grafico Exponencial
    ax1.stem(n, y1, label='x[n]', linefmt='purple')
    #Grafico Escalon Movimiento
    markerline, stemlines, baseline = ax1.stem(n, y2_shifted, '-.', label='h[n]', linefmt='green')
    
    ax1.set_xlim(-5, 15)
    ax1.set_ylim(0, 2)
    ax1.legend();
    ax1.set_title('Barrido Señales');

    #Grafico Convolución
    
    t2 = frame - 5

    ax2.stem(n,convpos(n,t2), label='h[n]', linefmt='red')
    

ani = animation.FuncAnimation(fig2, update, frames=np.arange(0, 25), interval=500)

# Mostrar la animación

HTML(ani.to_html5_video())
```
